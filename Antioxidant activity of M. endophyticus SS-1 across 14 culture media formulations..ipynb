## Result-Figure 1: Antioxidant activity of M. endophyticus SS-1 across 14 culture media formulations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load and clean data
df = pd.read_excel("14 media DPPH Scavanging test.xlsx")
df_renamed = df.rename(columns={"DPPH scavanging percentage": "DPPH (%)"})

# Sort by DPPH activity
df_sorted = df_renamed.sort_values("DPPH (%)", ascending=False)

# Create figure
plt.figure(figsize=(8,6))

# Create the bar plot
ax = sns.barplot(
    data=df_sorted,
    x="DPPH (%)",
    y="Media Name",
    palette="viridis"
)

# Adjust labels and title with added spacing
plt.xlabel("DPPH Scavenging Activity (%)", fontsize=14, labelpad=15)  # Adding space below x-label
plt.ylabel("Media Formulation", fontsize=14, labelpad=15)  # Adding space below y-label

# Adjust layout for better spacing
plt.tight_layout(pad=4.0)

# Save the figure with high resolution
plt.savefig("Figure1_DPPH_HorizontalBar.png", dpi=300, bbox_inches="tight")

# Show the plot
plt.show()

print("✅ Figure saved successfully as 'Figure1_DPPH_HorizontalBar.png'")
```
